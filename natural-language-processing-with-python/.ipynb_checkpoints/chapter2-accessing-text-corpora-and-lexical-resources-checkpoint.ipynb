### Gutenberg Corpus ###
+ [Project Gutenberg](http://www.gutenberg.org/) is an electronic text archive, which contains some 25,000 free electronic books.

```python
%matplotlib inline

import nltk
from nltk.corpus import gutenberg as gb

print("gutenberg fileids\n:{0}\n".format(gb.fileids()))
emma=gb.words("austen-emma.txt")
print("emma length:\n{0}\n".format(len(emma)))

# for concordance, we need to convert the words 
# to Text contrary to chapter 1
emma_text=nltk.Text(emma)
emma_text.concordance("surprise")

# This program displays three statistics for each text: average word length, average 
# sentence length, and the number of times each vocabulary item appears in the text 
# on average (our lexical diversity score). Observe that average word length appears 
# to be a general property of English, since it has a recurrent value of 4. (In fact, 
# the average word length is really 3 not 4, since the num_chars variable counts 
# space characters.) By contrast average sentence length and lexical diversity appear 
# to be characteristics of particular authors.
print("\n:average word length, average sentence length, lexical diversity, fileid")
for fileid in gb.fileids():
     num_chars = len(gb.raw(fileid))
     num_words = len(gb.words(fileid))
     num_sents = len(gb.sents(fileid))
     num_vocab = len(set(w.lower() for w in gb.words(fileid)))
     print(round(num_chars/num_words), round(num_words/num_sents), \
           round(num_words/num_vocab), fileid)
```

```python
# The sents() function divides the text up into its sentences, 
# where each sentence is a list of words
macbeth_sentences=gb.sents("shakespeare-macbeth.txt")
print("\nmacbeth_sentences:\n{0}\n".format(macbeth_sentences))
print("macbeth_sentences[1116]:\n{0}\n".format(macbeth_sentences[1116]))
longest_len=max(len(s) for s in macbeth_sentences)
print([s for s in macbeth_sentences if len(s) == longest_len])
```

```python
from nltk.corpus import brown

# categories
b_cat=brown.categories()
print("brown categories: {}".format(b_cat))

# words in a category
b_wrds_cat=brown.words(categories="news")
print("\nbrown words,categories:{}".format(b_wrds_cat))

b_wrds_fi=brown.words(fileids=["cg22"])
print("\nbrown words,fileids:{}".format(b_wrds_fi))

# sentences in categories
b_snts=brown.sents(categories=["news","editorial","reviews"])
print("\nbrown sentences:{}".format(b_snts))
```

```python
# Stylistics: study systematic differences between genres.
# model verbs: an auxiliary verb that expresses necessity or possibility, 
# such as 'can','could'
from nltk.corpus import brown
news_text=brown.words(categories="news")
fdist=nltk.FreqDist([w.lower() for w in news_text])
modals=["can","could","may","might","must","will"]
for m in modals:
    print("{}:{}".format(m,fdist[m]))
```

```python
# Obtain modal verb count for each genre of interest
cfd=nltk.ConditionalFreqDist(
            (genre,word)
            for genre in brown.categories()
            for word in brown.words(categories=genre))
genres=["news","religion","hobbies","science_fiction","romance","humor"]
cfd.tabulate(conditions=genres,samples=modals)

# Observe that the most frequent modal in the news genre is will, while 
# the most frequent modal in the romance genre is could. 
# Would you have predicted this? 
# The idea is that word counts might distinguish between genres.
```

Basic Corpus Functionality defined in NLTK: more documentation can be found using help(nltk.corpus.reader) and by reading the online Corpus HOWTO at http://www.nltk.org/howto.

## 2.2 Conditional Frequency Distributions ##

Given some list mylist of words or other items, FreqDist(mylist) would compute the number of occurrences of each item in the list.

A conditional frequency distribution (using NLTK's ConditionalFreqDist data type) is a collection of frequency distributions, each one for a different "condition". The condition will often be the category of the text.

### Conditions and Events ###

```python
# for FreqDist, we have counts of events such as appearance of words in a text
text = ['The', 'Fulton', 'County', 'Grand', 'Jury', 'said', ...]
# for conditional frequency distribution, we pair each event (word occurence) 
# with a condition (text category - genre).
# (condition, event)
pairs = [('news', 'The'), ('news', 'Fulton'), ('news', 'County'), ...]
```

### Plotting and Tabulating Distributions ###

In the plot() and tabulate() methods, we can optionally specify:
+ which conditions to display with a conditions= parameter. 
+ limit the samples to display with a samples= parameter. 

```python
from nltk.corpus import udhr
languages = ['Chickasaw', 'English', 'German_Deutsch',\
             'Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
cfd = nltk.ConditionalFreqDist(
           (lang, len(word))
           for lang in languages
           for word in udhr.words(lang + '-Latin1'))

# we can tabulate the cumulative frequency data just for two languages, 
# and for words less than 10 characters long
cfd.tabulate(conditions=['English','German_Deutsch'],samples=range(10),cumulative=True)

# We interpret a the last cell on the top row to mean that 1,638 words of the 
# English text have 9 or fewer letters.
```

### Functions ###


```python
def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2:] in ['sh', 'ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'

print(plural('fairy'))
print(plural('woman'))
```

### Wordlist Corpora ###

An example of the Words Corpus is the /usr/share/dict/words file from Unix, used by some spell checkers.

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

### Stopwords ###
High-frequency words like *the, to* and *also*.
Stopwords usually have little lexical content, and their presence in a text fails to distinguish it from other texts, hence we would like to filter them out before further processing.

```python
nltk.corpus.stopwords.words('english')
```

```python
# fraction of words in a text not in stopword list
def content_fraction(text):
    stopwords=nltk.corpus.stopwords.words('english')
    content=[w for w in text if w.lower() not in stopwords]
    return len(content)/len(text)

content_fraction(nltk.corpus.reuters.words())
```

### A Pronouncing Dictionary ###
NLTK includes the CMU Pronouncing Dictionary for US English, which was designed for use by speech synthesizers.

For each word, this lexicon provides a list of phonetic codes — distinct labels for each contrastive sound — known as *phones*.

```python
entries = nltk.corpus.cmudict.entries()
len(entries)
```

```python
for entry in entries[39943:39951]:
    print(entry)
```

```python
for word, pron in entries:
    if len(pron)==3:
        ph1,ph2,ph3=pron
        if ph1=='P' and ph3=='T':
            print(word,ph2)
```

```python
# Finds all words whose pronunciation ends with a syllable sounding like nicks. 
# You could use this method to find rhyming words.
syllable=['N','IH0','K','S']
print([word for word,pron in entries if pron[-4:]==syllable])
```

```python
print([w for w, pron in entries if pron[-1] == 'M' and w[-1] == 'n'])
```

```python
print(sorted(set(w[:2] for w, pron in entries if pron[0] == 'N' and w[0] != 'n')))
```

```python
# The phones contain digits to represent primary stress (1), secondary stress (2) 
# and no stress (0).
# we define a function to extract the stress digits and then scan our lexicon to find 
# words having a particular stress pattern.
def stress(pron):
     return [char for phone in pron for char in phone if char.isdigit()]
```

```python
print(([w for w,pron in entries if stress(pron)==['0','1','0','2','0']][:25]),'...')
```

```python
print([w for w,pron in entries if stress(pron)==['0', '2', '0', '1', '0']][:25],'...')
```

```python
# We can use a conditional frequency distribution to help us find minimally-contrasting 
# sets of words. Here we find all the p-words consisting of three sounds, and group them 
# according to their first and last sounds 
p3 = [(pron[0]+'-'+pron[2], word)
       for (word, pron) in entries
       if pron[0] == 'P' and len(pron) == 3]
cfd = nltk.ConditionalFreqDist(p3)
for template in cfd.conditions():
     if len(cfd[template]) > 10:
         words = cfd[template].keys()
         wordlist = ' '.join(words)
         print(template, wordlist[:70] + "...")
```

```python
def get_lemma_names(word):
    lemma_names_dict=dict()
    for synset in wn.synsets(word):
        lemma_names_dict[synset]=synset.lemma_names()
    return lemma_names_dict

print(get_lemma_names('car').values())
```
